```python
import numpy as np
import pandas as pd
```

```python
population = pd.read_csv('population.csv')

population = population.rename(index=str, columns={"Country (or dependent territory)": "Country"})
population.head()
```

```python
medals = pd.read_csv('medals.csv')
medals.head()
```

```python
results = pd.merge(population, medals, on="Country", how='left')
results = results.fillna(0)
results
```
