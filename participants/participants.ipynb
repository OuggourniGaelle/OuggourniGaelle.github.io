```python
import pandas as pd
```

```python
df = pd.read_csv('Olympics.tsv',sep='\t')
df.head()
```

```python
df = df[['Athlete_unique_url', 'Athlete','Trigram','Country','Event','Gender','Sport','Gold','Silver','Bronze','Total']]
df.head()
```

```python
results = df.groupby(
    ['Athlete_unique_url','Athlete','Trigram','Country','Event','Gender','Sport'], 
    as_index=False
)[['Gold','Silver','Bronze','Total']].sum()
```

```python
results
```

```python
results_count = df.groupby(
    ['Athlete_unique_url','Athlete','Trigram','Country','Event','Gender','Sport'], 
    as_index=False
).count()
```

```python
results_count
```

```python
results['Participations'] = results_count['Total']
```

```python
results
```

```python
for n in range (2,11):
    results[results['Participations']>n].to_csv("participants" + str(n) + ".csv", index=False, encoding='UTF-8')
    
```
