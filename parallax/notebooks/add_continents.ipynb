```python
data_path = './parallax/data/'
import pandas as pd
import numpy as np
import math
```

```python
countries = pd.read_csv(data_path + 'd3_bubbles.csv', na_filter = False)
countries.head()
```

```python
continents = countries[["CountryName", "ContinentCode"]]
continents = continents.rename(index=str, columns={"CountryName": "Country"})
continents
```

```python
#We add some missing countries
new_countries = pd.DataFrame([
        ['United States Virgin Islands', 'NA'], 
        ['Great Britain', 'EU'],
        ['Russia', 'EU'],
        ['Soviet Union', 'EU'],
        ['South Korea', 'AS'],
        ['Yugoslavia', 'EU'],
        ['Czechoslovakia', 'EU'],
        ['Chinese Taipei', 'AS'],
        ['East Germany','EU'],
        ['United States Virgin Islands', 'NA'],
        ['Libya', 'AS'],
        ['West Germany', 'EU'],
        ['North Korea', 'AS'],
        ['Federated States of Micronesia', 'OC']
    ], columns=['Country', 'ContinentCode'])
new_countries
```

```python
continents = continents.append(new_countries)
continents
```

```python
continents.reset_index(drop=True, inplace=True)
continents
```

```python
participants = pd.read_csv(data_path + 'participants3.csv')
participants.head()
```

```python
results = pd.merge(participants, continents, how='left', on=['Country'])
results.head()
```
