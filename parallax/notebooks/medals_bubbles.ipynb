```python
import pandas as pd
```

```python
#!ls ..
countries = pd.read_csv("countries.csv")
population_and_medals = pd.read_csv("population_and_medals.csv", encoding = "ISO-8859-1")
```

```python
print(countries.shape)
print(population_and_medals.shape)
```

```python
countries
```

```python
population_and_medals.head()
```

```python
data_to_add = population_and_medals[['medals']]
data_to_add['CountryName']=population_and_medals['Country']
data_to_add.head()
```

```python
results = pd.merge(countries, data_to_add, on="CountryName", how='left')
results['ContinentCode'] = results['ContinentCode'].fillna('NA')
results['medals'] = results['medals'].fillna(0)
results
```

```python
results.to_csv("d3_bubbles.csv", index=False, encoding='UTF-8')
```
